```python

import pylab
import scipy.stats as stats

indexlist_derog=df.derog.index.values.astype(int)
indexlist_delinq=df.delinq.index.values.astype(int)
delinq_bad= [df.delinq[i]  for i in indexlist_delinq if df.bad[i]==1]
delinq_good=[df.delinq[i]  for i in indexlist_delinq if df.bad[i]==0]
derog_bad= [df.derog[i]  for i in indexlist_derog if df.bad[i]==1]
derog_good= [df.derog[i]  for i in indexlist_derog if df.bad[i]==0]

#bu değişkenleri iyi ve kötü olarak ayırdım


```

```python
from scipy.stats import shapiro
print(shapiro(delinq_bad))
print(shapiro(delinq_good))
print(shapiro(derog_bad))
print(shapiro(derog_good))
#değişkenlerin normal dağılımdan gelip gelmediğine bakıyorum.
#hiçbiri normal dağılımdan gelmiyor ama değer sayıları 30 dan fazla olduğu için normal dağılıma yakınsadıklarını varsayıyorum
```

```python
from scipy.stats import levene 
print(levene(delinq_bad, delinq_good))
print(levene(derog_bad, derog_good))
```

iki testimizde de p değerimiz alfa = 0.05 ten küçük olduğu için H0 varyanslar eşittir hipotezini reddediyoruz.

```python
from scipy import stats
print(stats.ttest_ind(delinq_bad, delinq_good, equal_var = False))
print(stats.ttest_ind(derog_bad, derog_good, equal_var = False))
# H0 : Mu(delqbad) = Mu(delqgood)
# H1 : Mu(delqbad) != Mu(delqbad)
# p value alfadan küçük olduğu için reject H0
```

iki hipotezimizde rejecte düştü yani delinq ve derog değişkenleri kötü ve iyi müşterileri ayırmada iyi bir ölçüt değil (?)
